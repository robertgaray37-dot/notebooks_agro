#### **Celda 5: Código Python (Paso 4)**

```python
import pandas as pd
import numpy as np

# Simulamos un dataset de propiedades de materiales (Ej. Ingeniería Civil / Agroindustrial)
np.random.seed(42)
datos_ingenieria = pd.DataFrame({
    'ID_Muestra': range(1, 101),
    'Resistencia_MPa': np.random.normal(30, 5, 100),
    'Humedad_pct': np.random.uniform(10, 25, 100),
    'Categoria_Material': np.random.choice(['Tipo_A', 'Tipo_B'], 100)
})

# Introducimos algunos valores faltantes (NaN) intencionales para el ejercicio
datos_ingenieria.loc[10:14, 'Resistencia_MPa'] = np.nan
datos_ingenieria.loc[85:88, 'Humedad_pct'] = np.nan
```

```markdown
El siguiente código se encarga de identificar y cuantificar los valores faltantes en el DataFrame `datos_ingenieria`.

```python
print("\n--- 2. DETECCIÓN DE VALORES FALTANTES ---")
```
Esta línea imprime un encabezado para la sección, indicando el propósito de las operaciones siguientes: la detección de valores faltantes.

```python
faltantes = datos_ingenieria.isna().sum()
```
Aquí se realiza el proceso de detección de valores faltantes:
1.  `datos_ingenieria.isna()`: Este método de Pandas DataFrame devuelve un DataFrame booleano del mismo tamaño que `datos_ingenieria`. Cada celda contendrá `True` si el valor correspondiente en `datos_ingenieria` es un valor faltante (NaN, None, NaT) y `False` en caso contrario.
2.  `.sum()`: Al aplicar `.sum()` a este DataFrame booleano, Python interpreta `True` como 1 y `False` como 0. Por lo tanto, `sum()` calcula la cantidad de valores `True` (es decir, la cantidad de valores faltantes) por columna. El resultado se almacena en la variable `faltantes`, que será una Serie de Pandas donde el índice son los nombres de las columnas y los valores son los conteos de `NaN`.

```python
print(faltantes[faltantes > 0])
```
Esta línea filtra la Serie `faltantes` para mostrar solo aquellas columnas que tienen al menos un valor faltante. `faltantes > 0` crea una Serie booleana y al usarla para indexar `faltantes`, solo se seleccionan las filas (columnas originales) donde el conteo de faltantes es mayor que cero.

```python
# Interpretación: Saber cuántos faltan y en qué variables nos dice si debemos
# imputar datos o descartar registros antes de calcular estadísticas.
```
Este es un comentario explicativo que subraya la importancia de este paso: comprender la cantidad y ubicación de los valores faltantes es fundamental para tomar decisiones informadas sobre cómo manejarlos (imputarlos o eliminarlos) antes de proceder con análisis estadísticos más profundos.


```markdown
El siguiente código se enfoca en analizar la forma de la distribución de la variable 'Resistencia_MPa' mediante el cálculo de la asimetría (skewness) y la curtosis.

```python
print("\n--- 4. ASIMETRÍA (SKEWNESS) Y CURTOSIS ---")
```
Esta línea imprime un encabezado para la sección, indicando que el análisis se centrará en la asimetría y la curtosis.

```python
resistencia_limpia = datos_ingenieria['Resistencia_MPa'].dropna()
```
Aquí se crea una nueva Series llamada `resistencia_limpia`. Se selecciona la columna 'Resistencia_MPa' del DataFrame `datos_ingenieria` y se utiliza el método `.dropna()` para eliminar cualquier valor `NaN` (valores faltantes) de esta Series. Esto es crucial porque las funciones de asimetría y curtosis de `scipy.stats` no manejan automáticamente los valores faltantes y requerirían que los datos estén limpios.

```python
asimetria = skew(resistencia_limpia)
```
Se calcula el coeficiente de asimetría (skewness) de la distribución de `resistencia_limpia` utilizando la función `skew()` de la librería `scipy.stats`. La asimetría mide la simetría de la distribución. Un valor cercano a 0 indica simetría, valores positivos indican una cola más larga a la derecha (asimetría positiva), y valores negativos indican una cola más larga a la izquierda (asimetría negativa).

```python
curtosis_val = kurtosis(resistencia_limpia)
```
Se calcula el coeficiente de curtosis de la distribución de `resistencia_limpia` utilizando la función `kurtosis()` de la librería `scipy.stats`. La curtosis mide la 'puntiagudez' de la distribución y el peso de sus colas. Un valor de curtosis cercano a 0 (para la curtosis Fisher, que es la predeterminada en `scipy.stats`) indica una distribución mesocúrtica (similar a una distribución normal). Valores positivos indican una distribución leptocúrtica (más puntiaguda y con colas más pesadas), mientras que valores negativos indican una distribución platicúrtica (más plana y con colas más ligeras).

```python
print(f"Asimetría de la Resistencia: {asimetria:.2f}")
print(f"Curtosis de la Resistencia: {curtosis_val:.2f}")
```
Estas líneas imprimen los valores calculados de asimetría y curtosis para la 'Resistencia_MPa', formateados a dos decimales para una mejor legibilidad.

```python
# Interpretación Ingenieril:
# - Asimetría cercana a 0 indica distribución simétrica.
# - Curtosis > 0 indica colas pesadas (mayor probabilidad de valores extremos o fallos inesperados en el material).
```
Este es un comentario explicativo que proporciona una guía sobre cómo interpretar los valores de asimetría y curtosis desde una perspectiva de ingeniería. Destaca que una asimetría cercana a cero sugiere simetría, y una curtosis positiva podría indicar una mayor probabilidad de eventos extremos o fallos en el material.

````
# ==========================================
# PASO 4: FORMA DE LA DISTRIBUCIÓN
# ==========================================
print("\n--- 4. ASIMETRÍA (SKEWNESS) Y CURTOSIS ---")

# Calculamos asimetría y curtosis solo para variables numéricas, omitiendo los NaN
resistencia_limpia = datos_ingenieria['Resistencia_MPa'].dropna()

asimetria = skew(resistencia_limpia)
curtosis_val = kurtosis(resistencia_limpia)

print(f"Asimetría de la Resistencia: {asimetria:.2f}")
print(f"Curtosis de la Resistencia: {curtosis_val:.2f}")

# Interpretación Ingenieril:
# - Asimetría cercana a 0 indica distribución simétrica.
# - Curtosis > 0 indica colas pesadas (mayor probabilidad de valores extremos o fallos inesperados en el material).
````
