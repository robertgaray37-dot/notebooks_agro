```r
# Instalar librería si no está disponible
# install.packages("moments")
library(moments)

# 1. Recrear un dataframe básico con valores faltantes (NA)
set.seed(42)
datos_r <- data.frame(
  Resistencia = rnorm(100, 30, 5),
  Humedad = runif(100, 10, 25),
  Categoria = sample(c('Tipo_A', 'Tipo_B'), 100, replace=TRUE)
)
datos_r$Resistencia[11:15] <- NA

# PASO 1: Inspección estructural
print("Estructura:")
str(datos_r)

# PASO 2: Valores Faltantes
print("Valores NA por columna:")
colSums(is.na(datos_r))

# PASO 3: Análisis Univariado
print("Resumen Estadístico:")
summary(datos_r)

# PASO 4: Forma de la distribución (omitiendo NA)
resistencia_limpia <- na.omit(datos_r$Resistencia)
print(paste("Asimetría (Skewness):", skewness(resistencia_limpia)))
print(paste("Curtosis:", kurtosis(resistencia_limpia)))
```
